```python
import warnings
import numpy as np
import pandas as pd
import pyfixest as pf
import seaborn as sns
import matplotlib.pyplot as plt
import pyreadstat
from statsmodels.nonparametric.smoothers_lowess import lowess
import statsmodels.formula.api as smf
from typing import List
import pyreadstat

wage_data, meta = pyreadstat.read_dta("L15_wage_postgrad_clean.dta")
```

## 1)

```python
reg_w_f = pf.feols("lnw~female", data=wage_data,vcov="HC1")
reg_w_fa = pf.feols("lnw~female+age", data=wage_data,vcov="HC1")
reg_w_faa = pf.feols("lnw~female+age+agesq", data=wage_data,vcov="HC1")
reg_w_faaa = pf.feols("lnw~female+age+agesq+agecu+agequ", data=wage_data,vcov="HC1")
```

```python
pf.etable([reg_w_f, reg_w_fa, reg_w_faa, reg_w_faaa],
          head_order="h",
          model_heads=["ln wage","ln wage", "ln wage", "ln wage"],
          labels={"Intercept": "Constant"}          
)
```

## 2)

```python
reg_w_f = pf.feols("lnw~female", data=wage_data,vcov="HC1")
reg_w_fpph= pf.feols("lnw~female+edProf+edPhd", data=wage_data,vcov="HC1")
reg_w_fpm = pf.feols("lnw~female+edProf+edMA", data=wage_data,vcov="HC1")
reg_w_fphm = pf.feols("lnw~female+edPhd+edMA", data=wage_data,vcov="HC1")
```

```python
pf.etable([reg_w_f, reg_w_fpph, reg_w_fpm, reg_w_fphm],
          head_order="h",
          model_heads=["ln wage","ln wage", "ln wage", "ln wage"],
          labels={"Intercept": "Constant"}          
)
```

## 3)

```python
wage_data_res = wage_data.loc[(wage_data['age']>=40) & (wage_data['age']<=60)].copy()
```

```python
reg_w_f_res = pf.feols("lnw~female", data=wage_data_res,vcov="HC1")
reg_w_fpph_res = pf.feols("lnw~female*edMA", data=wage_data_res,vcov="HC1")
reg_w_fpm_res = pf.feols("lnw~female*edPhd", data=wage_data_res,vcov="HC1")
reg_w_fphm_res = pf.feols("lnw~female*edProf", data=wage_data_res,vcov="HC1")
```

```python
pf.etable([reg_w_f_res, reg_w_fpph_res, reg_w_fpm_res, reg_w_fphm_res],
          head_order="h",
          model_heads=["ln wage","ln wage", "ln wage", "ln wage"],
          labels={"Intercept": "Constant"}          
)
```
